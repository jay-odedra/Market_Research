Loading in modules

```python

import datetime as dt
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.offline as pyo
import plotly.graph_objects as go
from plotly.subplots import make_subplots
pyo.init_notebook_mode(connected=True)
pd.options.plotting.backend = "plotly"
```

```python
end = dt.datetime(2020, 12, 31)
start = end - dt.timedelta(days=3650)
```

```python
stocklist = ["BARC", "HSBA", "LLOY", "NWG"]
stocklist = [stock + ".L" for stock in stocklist]
```

```python
df = yf.download(stocklist, start=start, end=end).stack(future_stack=True)
df
```

```python
df
```

```python
df_barc = df.xs("BARC.L", level=1)
df_barc
```

```python
df_barc= df_barc.copy()
df_barc["Daily_Return"] = df_barc["Close"] - df_barc["Open"]
df_barc["Daily_Return_Percent"] = (df_barc["Close"] - df_barc["Open"]) / df_barc["Open"] * 100

df_barc["Max_change"] = (df_barc["High"] - df_barc["Low"])
df_barc["Max_change_Percent"] = (df_barc["High"] - df_barc["Low"]) / df_barc["Low"] * 100

df_barc['Open_Close_Diff'] = df_barc['Open'] - df_barc['Close'].shift(1)
df_barc
```

```python
df_barc['MA50'] = df_barc['Close'].rolling(window=50, min_periods=0).mean()
df_barc['MA200'] = df_barc['Close'].rolling(window=200, min_periods=0).mean()
df_barc['MA200'].head(30)
```

```python
fig = go.Figure()

fig.add_trace(go.Candlestick(x=df_barc.index, open=df_barc["Open"], high=df_barc["High"],
                low=df_barc["Low"], close=df_barc["Close"], name="OHLC"))

fig.update_layout(
    title='BARC historical price chart',
    xaxis_tickfont_size=12,
    yaxis=dict(
        title='Share Price (Â£))',
        tickfont_size=12,
    ),
    autosize=False,
    width=1600,
    height=1000,
    margin=dict(l=50, r=50, b=50, t=50, pad=2),
    paper_bgcolor='white'
)

fig.show()
```

```python
fig.add_trace(go.Scatter(x=df_barc.index, y=df_barc["MA50"], marker_color='grey',name="MA50"))
fig.add_trace(go.Scatter(x=df_barc.index, y=df_barc["MA200"], marker_color='black',name="MA200"))

fig.show()
```

```python
fig = go.Figure(data=[go.Histogram(x=df_barc['Open_Close_Diff'], nbinsx=500, histnorm='probability')])
fig.update_layout(title_text='Open Close Difference Histogram', xaxis_title_text='Open Close Difference', yaxis_title_text='Probability')
fig.update_xaxes(range=[-15,15])

fig.show()
```
