```python
df = get_stock_dataset('BARC.L', end_date=[2024, 1, 1], days=3650)
df
```

```python
log_return = get_return_Ndays(df, 1, log=True)
log_return
```

```python
daily_std = log_return.std()
daily_std
```

```python
annualized_vol = daily_std * np.sqrt(252)
annualized_vol*100
```

```python
trading_days = 60
volatility = log_return.rolling(window=trading_days,min_periods=1).std() * np.sqrt(trading_days)
volatility.plot().update_layout(autosize=True)
```

```python
Rf = 0.01/252
sharpe_ratio = (log_return.rolling(window=trading_days,min_periods=10).mean() - Rf)*trading_days / volatility
```

```python
sharpe_ratio.plot().update_layout(autosize = True).show()
```

```python
sortino_vol = log_return[log_return<0].rolling(window=trading_days,center=True,min_periods=10).std() * np.sqrt(trading_days)
sortino_ratio = (log_return.rolling(window=trading_days).mean() - Rf)*trading_days / sortino_vol
sortino_vol.plot().update_layout(autosize = False, width=600, height=300).show()
sortino_ratio.plot().update_layout(autosize = False, width=600, height=300).show()
```
